```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re


```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
  text = str(text).lower()
  # reomved the punctuations
  text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
  words = text.split()
  # remove the stopword
  words = [word for word in words if word not in stop_words]
  #lemmatization
  words = [lemmatizer.lemmatize(word) for word in words]
  return ' '.join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df['tags'][1]
```

```python
df = df.reset_index(drop = True)
```

```python
indices = pd.Series(df.index,index  = df['title']).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words = 'english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix.shape
```

```python
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
def recommend(title,n = 10):
  if title not in indices:
    return ['Movie not found']

  idx = indices[title]
  similarity_scores = cosine_similarity(tfidf_matrix[idx],tfidf_matrix)
  similarity_scores = list(enumerate(similarity_scores[0]))
  similarity_scores = sorted(similarity_scores,key = lambda x:x[1],reverse = True)[1:n+1]

  recommended_movies = [df['title'].iloc[i[0]] for i in similarity_scores]
  return recommended_movies
```

```python
recommend('Avenger')
```

```python
import pickle

pickle.dump(tfidf_matrix,open('tfidf_matrix.pkl','wb'))
pickle.dump(indices,open('indices.pkl','wb'))

df.to_pickle('df.pkl')

pickle.dump(tfidf,open('tfidf.pkl','wb'))
```
